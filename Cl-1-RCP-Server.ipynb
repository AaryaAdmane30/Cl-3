```python
from xmlrpc.server import SimpleXMLRPCServer
# # Python has a built-in module called xmlrpc.
# SimpleXMLRPCServer helps create a server that can communicate over a network using the XML-RPC protocol.
# What is XML-RPC?

# XML-RPC is a protocol where:

# data is sent in XML format
# communication happens using HTTP

# Basically:

# A client calls a function on another computer like it’s local.

# Ancient but still important for understanding distributed systems and RPC concepts.
```

```python
# Create server
server = SimpleXMLRPCServer(("localhost", 8000))
print("Server is running on port 8000...")

# Create Server
# server = SimpleXMLRPCServer(("localhost", 8000))

# This creates the XML-RPC server.

# Meaning:
# "localhost" → server runs on your own computer
# 8000 → port number

# So the server address becomes:

# http://localhost:8000
```

```python

# Register function
server.register_function(factorial, "factorial")  # This makes the factorial() function available to clients.
```

```python
# Run server
server.serve_forever()
```
