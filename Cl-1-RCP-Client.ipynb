```python
import xmlrpc.client

#  It allows your program to:

# connect to a remote XML-RPC server
# call functions on that server

#  Hey Python, I want to communicate with another machine/program.”
```

```python
# Connect to server
client = xmlrpc.client.ServerProxy("http://localhost:8000/")

#  Server was created using:

# SimpleXMLRPCServer(("localhost", 8000))

# So client must connect to the same port.


# What is ServerProxy?

# ServerProxy acts like a bridge between:

# client program
# remote server

# After connection:

# client.factorial()

# That’s the whole RPC concept:Call remote functions like local functions.
```

```python
# Input from user
num = int(input("Enter a number: "))

```

```python
# RPC call
result = client.factorial(num)

```

```python
# Output result
print("Factorial is:", result)
```

```python
# Client sends request
#         ↓
# Server receives request
#         ↓
# Server runs factorial(5)
#         ↓
# Server returns 120
#         ↓
# Client stores result


# # +------------+          Network          +------------+
# |   Client   |  --------------------->  |   Server   |
# |            |   factorial(5) request   |            |
# |            |  <---------------------  | factorial()|
# | gets 120   |        returns 120       | calculates |
# +------------+                          +------------+
```

```python
# RPC is Remote Procedure Call. It is used for client-server applications. RPC mechanisms are 
# used when a computer program causes a procedure or subroutine to execute in a different 
# address space, which is coded as a normal procedure call without the programmer specifically 
# coding the details for the remote interaction. 
# This procedure call also manages low-level transport protocol, such as User Datagram 
# Protocol, Transmission Control Protocol/Internet Protocol etc. It is used for carrying the 
# message data between programs. 
# Types of RPC 
# Three types of RPC are: 
# • Callback RPC 
# • Broadcast RPC 
# • Batch-mode RPC 
# Callback RPC 
# This type of RPC enables a P2P paradigm between participating processes. It helps a process to 
# be both client and server services. 
# Functions of Callback RPC: 
# • Remotely processed interactive application problems 
# • Offers server with clients handle 
# • Callback makes the client process wait 
# • Manage callback deadlocks 
# • It facilitates a peer-to-Peer paradigm among participating processes. 
# Broadcast RPC 
# Broadcast RPC is a client’s request, that is broadcast on the network, processed by all servers 
# which have the method for processing that request. 
# 6
# Department of Artificial Intelligence and Data Science, DYPIEMR, Pune 
# [2023-24] 
# Computer Laboratory-III                               
# B.E.(Sem-II)    
# Functions of Broadcast RPC: 
# • Allows you to specify that the client’s request message has to be broadcasted. 
# • You can declare broadcast ports. 
# • It helps to reduce the load on the physical network 
# Batch-mode RPC 
# Batch-mode RPC helps to queue, separate RPC requests, in a transmission buffer, on the client
# side, and then send them on a network in one batch to the server. 
# Functions of Batch-mode RPC: 
# • It minimizes overhead involved in sending a request as it sends them over the network in 
# one batch to the server. 
# • This type of RPC protocol is only efficient for the application that needs lower call rates. 
# • It needs a reliable transmission protocol. 
# RPC Architecture 
# RPC architecture has mainly five components of the program: 
# 1. Client 
# 2. Client Stub 
# 3. RPC Runtime 
# 4. Server Stub 
# 5. Server
```
