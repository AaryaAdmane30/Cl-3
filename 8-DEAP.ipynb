```python
import random
import numpy as np
from deap import base, creator, tools, algorithms
import multiprocessing
from multiprocessing.pool import ThreadPool as Pool


# DEAP stands for:

# Distributed Evolutionary Algorithms in Python

# It is a Python framework used to implement:

# Genetic Algorithms
# Evolutionary Algorithms
# Optimization techniques

# DEAP helps in solving difficult optimization problems using concepts inspired by natural evolution.

# 3. What is an Evolutionary Algorithm?

# An evolutionary algorithm is an optimization technique based on:

# Natural selection
# Mutation
# Survival of the fittest

# It works by:

# Creating random solutions
# Evaluating their fitness
# Selecting the best solutions
# Applying crossover and mutation
# Repeating the process for many generations
# The goal is to find the best possible solution.


# Why Distributed Evolutionary Algorithm?

# Complex optimization problems require high computation.

# Instead of using a single processor:

# The task is divided into multiple islands/nodes
# Each node evolves solutions independently
# Best solutions are exchanged between nodes

# Advantages:

# Faster execution
# Better optimization
# Handles large-scale problems


# Improve optimization using parallel processing.

# Meaning:

# Multiple islands work simultaneously
# Reduces computation time
# Improves solution quality



# Important Features of DEAP
# 1. Genetic Operators

# DEAP provides:

# Crossover
# Mutation
# Selection

# # #  deap → provides genetic algorithm tools.
# multiprocessing and Pool → run multiple islands in parallel.
```

```python
DIMENSIONS = 7
BOUNDS = (-5, 5)

#  Each solution has 7 variables. - Values can range from -5 to 5.

# Fitness Function
# his calculates the sum of squares of all values. - f(x)=x12​+x22​+x32​+ [0,0,0,0,0,0,0]
# because fitness becomes 0.
def fitness_function(individual):
    return (sum(x**2 for x in individual),)
```

```python

# Creating Fitness and Individual
# FitnessMin → tells GA to minimize fitness.
# Individual → defines each chromosome as a list.
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)

# Toolbox stores all GA operations.
toolbox = base.Toolbox()
# Creating Population
toolbox.register("attr_float", random.uniform, BOUNDS[0], BOUNDS[1])

# Creates random numbers between -5 and 5.
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, DIMENSIONS)
# Creates one individual with 7 random values.
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Creates a population of individuals.

# Genetic Operators
# Evaluation
toolbox.register("evaluate", fitness_function)
# Combines two parents to create children. -
# Crossover
toolbox.register("mate", tools.cxBlend, alpha=0.5)
#  Mutation :
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)

# Adds random changes using Gaussian mutation.
# mu=0 → mean
# sigma=1 → standard deviation
# indpb=0.2 → mutation probability

#  selection : Uses tournament selection to choose best individuals.
toolbox.register("select", tools.selTournament, tournsize=3)
```

```python
def evolve_island(seed, generations=5):
    random.seed(seed)

    pop = toolbox.population(n=50)
    hof = tools.HallOfFame(1)
    log = []

    for gen in range(generations):
        offspring = algorithms.varAnd(pop, toolbox, cxpb=0.7, mutpb=0.2)
        fits = list(map(toolbox.evaluate, offspring))

        for ind, fit in zip(offspring, fits):
            ind.fitness.values = fit

        pop = toolbox.select(offspring, k=len(pop))
        hof.update(pop)

        best = hof[0]
        log.append((gen, best.fitness.values[0]))

    return pop, hof[0], log
```

```python
def migrate(islands):
    islands = [list(i) for i in islands]

    for i in range(len(islands)):
        next_i = (i + 1) % len(islands)

        best_individual = islands[i][1]

        # Replace worst individual in next island
        worst_idx = max(
            range(len(islands[next_i][0])),
            key=lambda j: islands[next_i][0][j].fitness.values[0]
        )

        islands[next_i][0][worst_idx] = best_individual

    return islands
```

```python
NUM_ISLANDS = 2

pool = Pool(processes=NUM_ISLANDS)

# Run islands in parallel
results = pool.map(evolve_island, range(NUM_ISLANDS))
results = [list(r) for r in results]

for i, (_, best, log) in enumerate(results):
    print(f"\nIsland {i} Log:")
    for gen, fit in log:
        print(f"Gen {gen+1} | Fitness: {fit:.4f}")

# Migration step
results = migrate(results)

# Get global best
best = min([res[1] for res in results], key=lambda ind: ind.fitness.values[0])

print("\nBest Individual:", [round(x, 4) for x in best])
print("Best Fitness:", best.fitness.values[0])
```
